```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filepath = r"C:\Users\KIIT\OneDrive\Desktop\Cleaned_dataset_Churn_Analysis.xlsx"
df = pd.read_excel(filepath)
```

```python

sns.set(style='whitegrid')
```

```python
print("Loaded rows:", len(df))
print("Columns:", list(df.columns))
```

```python
pd.set_option('display.max_columns', None)  # display all columns
pd.set_option('display.width', None)        # auto-adjust width
pd.set_option('display.max_colwidth', None) # show full content if long
df.head(5)
```

```python
# 1. Churn rate by Tenure Group
# ---------------------------
plt.figure(figsize=(8,5))
tenure_churn = df.groupby("tenure_group")["Churn_Flag"].mean().reset_index()
sns.barplot(x="tenure_group", y="Churn_Flag", data=tenure_churn, palette="viridis")
plt.title("Churn Rate by Tenure Group")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# 2. Churn by Contract Type

plt.figure(figsize=(6,4))
contract_churn = df.groupby("Contract")["Churn_Flag"].mean().reset_index()
sns.barplot(x="Contract", y="Churn_Flag", data=contract_churn, palette="magma")
plt.title("Churn Rate by Contract Type")
plt.ylabel("Churn Rate")
plt.show()
```

```python
# 3. Revenue vs Churn
# ---------------------------
plt.figure(figsize=(8,5))
sns.boxplot(x="Churn_Flag", y="MonthlyCharges", data=df, palette="Set2")
plt.title("Monthly Charges vs Churn")
plt.xticks([0,1], ["No Churn","Churn"])
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x="Churn_Flag", y="TotalCharges", data=df, palette="Set1")
plt.title("Total Charges vs Churn")
plt.xticks([0,1], ["No Churn","Churn"])
plt.show()
```

```python
# 4. Service Adoption Correlation with Churn
# ---------------------------
service_cols = ["OnlineSecurity_Flag", "OnlineBackup_Flag", "DeviceProtection_Flag",
                "TechSupport_Flag", "StreamingTV_Flag", "StreamingMovies_Flag"]

plt.figure(figsize=(8,6))
corr = df[service_cols + ["Churn_Flag"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation of Services with Churn")
plt.show()
```

```python
# 5. Comparative Plots (Tenure + Contract + Services)
# ---------------------------
plt.figure(figsize=(10,6))
sns.boxplot(x="Contract", y="MonthlyCharges", hue="Churn_Flag", data=df, palette="husl")
plt.title("Contract Type & Monthly Charges vs Churn")
plt.ylabel("Monthly Charges")
plt.show()
```
